```python
def isPalindrome(num):
    cnum = num 
    reverse = 0
    while cnum > 0:
        digit = cnum % 10
        reverse = reverse * 10 + digit 
        cnum = cnum // 10

    return num == reverse 

print(isPalindrome(121))
print(isPalindrome(1221))
print(isPalindrome(123)) 
```

```python
def isPalindrome(num):
    num_str = str(num) 
    left = 0
    right = len(num_str) - 1
    isThat = True 
    while left < right:
        if num_str[left] != num_str[right]:
            isThat = False 
            break 
        left += 1
        right -= 1
    return isThat 

print(isPalindrome(121))
print(isPalindrome(1221))
print(isPalindrome(123))
```

```python
def isPalindrome(num):
    cnum = num 
    reverse = 0
    while cnum > reverse:
        digit = cnum % 10
        reverse = reverse * 10 + digit 
        cnum = cnum // 10

    return cnum == reverse or cnum == reverse // 10

print(isPalindrome(121))    
print(isPalindrome(1221))
print(isPalindrome(1234)) 
print(isPalindrome(123)) 
```

```python
def countPrime(n): 
    def isPrime(num):
        if num <= 1:
            return False 
        limit = int(num ** 0.5)
        for I in range(2,limit + 1):
            if num % I == 0:
                return False         
        return True 
    count = 0
    for I in range(1,n):
        if isPrime(I):
            count += 1
    return count 

print(countPrime(100))
```

```python

def getPrimes(n): 
    def isPrime(num):
        if num <= 1:
            return False 
        limit = int(num ** 0.5)
        for I in range(2,limit + 1):
            if num % I == 0:
                return False         
        return True 
    primes = [num for num in range(1,n) if isPrime(num)]
    return primes 

print(getPrimes(100))
```
