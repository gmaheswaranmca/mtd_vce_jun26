```python
def findMaxProfit(prices):
    max_profit = 0
    n = len(prices)
    for I in range(n-1):
        for J in range(I+1, n):
            profit = prices[I] - prices[J]
            if profit > max_profit:
                max_profit = profit 
    return max_profit 

print(findMaxProfit([7, 1, 5, 3, 6, 4])) 
```
