```python
def two_sum(nums, target):
    n = len(nums)
    for I in range(n - 1):
        for J in range(I + 1, n):
            s = nums[I] + nums[J]
            if s == target:
                return [I, J]
    return [-1, -1]

print(two_sum([2,7,11,15], 18))
```

```python
def two_sum(nums, target):
    d = {}
    for index, num in enumerate(nums):
        diff = target - num 
        if diff in d:
            return [d[diff], index]
        d[num] = index 

    return [-1, -1]

print(two_sum([2,7,11,15], 18))
```
