```python
# Learning list
# Indexed collection
# Allows duplicates
# Can store different data types
# Maintains insertion order
# Readable (Immutable) 
nums = [10, 20, 10, 30, 40, 50, 10]

# Basic information
print(f'nums: {nums}')
print(f'number at index 1 = {nums[1]}, index 4 = {nums[4]}')

# Append elements
nums.extend([70, 80, 90])
print(f'after adding 70,80,90: {nums}')

# Pop last element
removed = nums.pop()
print(f'popped element: {removed}')
print(f'after pop: {nums}')

# Pop at specific index
removed = nums.pop(5)
print(f'popped at index 5: {removed}')
print(f'after pop(5): {nums}')

# Insert element
nums.insert(2, 25)
print(f'after insert 25 at index 2: {nums}')

# Extend list
nums.extend([100, 200, 300])
print(f'after extend: {nums}')

# Remove first occurrence of 10
if 10 in nums:
    nums.remove(10)
print(f'after remove(10): {nums}')

# Length
print(f'count elements: {len(nums)}')

# Count occurrences
print(f'count of 10: {nums.count(10)}')

# Index search
if 40 in nums:
    print(f'40 found at index {nums.index(40)}')

# List comprehension
ten_divisible = [num for num in nums if num % 10 == 0]
print(f'ten divisible: {ten_divisible}')

# Reverse
print(f'reversed list: {list(reversed(nums))}')

# Sorted copy
print(f'sorted list: {sorted(nums)}')

# Maximum and Minimum
print(f'max = {max(nums)}')
print(f'min = {min(nums)}')

# Sum
print(f'sum = {sum(nums)}')
```

```python
# Learning Tuple
# Indexed collection
# Allows duplicates
# Can store different data types
# Maintains insertion order
# Readable (Immutable)

nums_tuple = (10, 20, 10, 30, 40, 50, 10)

print(f'nums_tuple: {nums_tuple}')

# Accessing elements
print(f'number at index 1 = {nums_tuple[1]}')
print(f'number at index 4 = {nums_tuple[4]}')

# Length
print(f'total elements: {len(nums_tuple)}')

# Count occurrences
print(f'count of 10: {nums_tuple.count(10)}')

# Find index
print(f'first occurrence of 40 at index: {nums_tuple.index(40)}')

# Membership test
print(f'Is 50 present? {50 in nums_tuple}')
print(f'Is 100 present? {100 in nums_tuple}')

# Slicing
print(f'first 3 elements: {nums_tuple[:3]}')
print(f'last 3 elements: {nums_tuple[-3:]}')

# Iteration
print('Elements:')
for num in nums_tuple:
    print(num, end=' ')
print()

# Unpacking
a, b, c, d, e, f, g = nums_tuple
print(f'unpacked values: {a}, {b}, {c}, {d}, {e}, {f}, {g}')

# Convert tuple to list
nums_list = list(nums_tuple)
nums_list.append(60)

# Convert back to tuple
updated_tuple = tuple(nums_list)
print(f'updated tuple: {updated_tuple}')

# One element tuple
name_only = ('Mahesh',)
print(f'one element tuple: {name_only}')
print(f'type: {type(name_only)}')

# Without comma, it is not a tuple
name = ('Mahesh')
print(f'not a tuple: {name}')
print(f'type: {type(name)}')
```

```python
# Learning Set
# Does not allow duplicates
# Elements can be of different types
# Unordered collection
# Mutable (readable and writable)
# Fast lookup

nums_set = {10, 20, 10, 30, 40, 50, 10}

print(f'nums_set: {nums_set}')

# Length
print(f'total elements: {len(nums_set)}')

# Add element
nums_set.add(70)
print(f'after add(70): {nums_set}')

# Add multiple elements
nums_set.update([80, 90, 100])
print(f'after update([80,90,100]): {nums_set}')

# Membership testing
print(f'Is 40 present? {40 in nums_set}')
print(f'Is 500 present? {500 in nums_set}')

# Remove element
nums_set.remove(20)
print(f'after remove(20): {nums_set}')

# Discard (safe remove)
nums_set.discard(999)  # No error
print(f'after discard(999): {nums_set}')

# Pop random element
removed = nums_set.pop()
print(f'popped element: {removed}')
print(f'after pop(): {nums_set}')

# Traversal
print('Elements:')
for num in nums_set:
    print(num, end=' ')
print()

# Set Operations
set1 = {1, 2, 3, 4}
set2 = {4, 3, 5, 6}

print(f'set1: {set1}')
print(f'set2: {set2}')

# Union
print(f'Union: {set1 | set2}')

# Intersection
print(f'Intersection: {set1 & set2}')

# Difference
print(f'Difference (set1 - set2): {set1 - set2}')

# Symmetric Difference
print(f'Symmetric Difference: {set1 ^ set2}')

# Subset and Superset

print(f'set[1,2] subset of {set1} : {set([1,2]) <= set1}')
print(f'{set1} superset of set[1,2]? : {set1 >= {1,2}}')

# Clear all elements
temp_set = {1, 2, 3}
temp_set.clear()
print(f'after clear(): {temp_set}')
```

```python
# Learning Dictionary
# Key-Value pairs
# Keys must be unique
# Values can be duplicated
# Mutable (readable and writable)
# Preserves insertion order (Python 3.7+)

person = {
    'id': 101,
    'name': 'Mahesh',
    'age': 25,
    'city': 'Warangal'
}

print(f'person: {person}')

# Access values
print(f"id: {person['id']}")
print(f"name: {person['name']}")

# Safe access
print(f"salary: {person.get('salary', 'Not Available')}")

# Add new key
person['salary'] = 50000
print(f'after adding salary: {person}')

# Update existing key
person['age'] = 26
print(f'after updating age: {person}')

# Add multiple key-value pairs
person.update({
    'email': 'mahesh@example.com',
    'country': 'India'
})
print(f'after update(): {person}')

# Remove a key
removed_salary = person.pop('salary')
print(f'removed salary: {removed_salary}')
print(f'after pop(): {person}')

# Length
print(f'total keys: {len(person)}')

# Membership testing
print(f"'name' exists? {'name' in person}")
print(f"'salary' exists? {'salary' in person}")

# Keys
print(f'keys: {list(person.keys())}')

# Values
print(f'values: {list(person.values())}')

# Items
print(f'items: {list(person.items())}')

# Iterate keys
print('\nKeys:')
for key in person:
    print(key)

# Iterate values
print('\nValues:')
for value in person.values():
    print(value)

# Iterate key-value pairs
print('\nKey-Value Pairs:')
for key, value in person.items():
    print(f'{key} -> {value}')

# Dictionary Comprehension
square_dict = {num: num * num for num in range(1, 6)}
print(f'\nsquare_dict: {square_dict}')

# Copy dictionary
person_copy = person.copy()
print(f'copy: {person_copy}')

# Clear dictionary
temp = {'a': 1, 'b': 2}
temp.clear()
print(f'after clear(): {temp}')
```

```python
# Learning String
# Ordered sequence of characters
# Allows duplicates
# Immutable (read-only)
# Supports indexing and slicing

name = "Mahesh Kumar"

print(f'name: {name}')

# Length
print(f'length: {len(name)}')

# Indexing
print(f'first character: {name[0]}')
print(f'last character: {name[-1]}')

# Slicing
print(f'first 6 chars: {name[:6]}')
print(f'last 5 chars: {name[-5:]}')
print(f'every 2nd char: {name[::2]}')

# Case conversion
print(f'upper: {name.upper()}')
print(f'lower: {name.lower()}')
print(f'title: {name.title()}')

# Searching
print(f'contains "Kumar"? {"Kumar" in name}')
print(f'index of "Kumar": {name.find("Kumar")}')

# Replace
print(f'replace Kumar with Raj: {name.replace("Kumar", "Raj")}')

# Count
print(f'count of a: {name.count("a")}')

# Starts and Ends
print(f'starts with Mah? {name.startswith("Mah")}')
print(f'ends with Kumar? {name.endswith("Kumar")}')

# Split
words = name.split()
print(f'split: {words}')

# Join
joined = '-'.join(words)
print(f'joined: {joined}')

# Remove spaces
text = "   Python Programming   "
print(f'before strip: "{text}"')
print(f'after strip: "{text.strip()}"')

# String validation
print(f'"123".isdigit(): {"123".isdigit()}')
print(f'"abc".isalpha(): {"abc".isalpha()}')
print(f'"abc123".isalnum(): {"abc123".isalnum()}')

# Iteration
print('Characters:')
for ch in name:
    print(ch, end=' ')
print()

# String formatting
age = 25
print(f'My name is {name} and age is {age}')
```

```python
# Comprehension
# General syntax:
# [result_expression for item in collection]

# Comprehension
nums = [1, 2, 3, 4, 5]
squares = [num * num for num in nums]
print(squares)

# List Comprehension with Condition
nums = [1, 2, 3, 4, 5, 6, 7, 8]
evens = [num for num in nums if num % 2 == 0]
print(evens)
odds = [num for num in nums if num % 2 != 0]
print(odds)

# Set Comprehension
nums = [1, 2, 2, 3, 3, 4, 5]
unique_squares = {num * num for num in nums}
print(unique_squares)

# Dictionary Comprehension
nums = [1, 2, 3, 4, 5]
square_dict = {num: num * num for num in nums}
print(square_dict)

# String from Comprehension
name = "Mahesh"
vowels = ''.join(
    ch for ch in name
    if ch.lower() in 'aeiou'
)
print(vowels)
```
